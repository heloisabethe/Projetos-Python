```python
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
# Agrupar os dados por artista e calcular a média de popularidade
# dados_agrupados = dados_spotify.groupby('artists')['popularity'].mean().reset_index()- quando quero soment uma opção
agrupado_por_artista = dados_spotify.groupby('artists').agg({
    'popularity': 'mean',  # Média de popularidade
    'track_id': 'count',   # Número de faixas
    'duration_ms': 'mean', # Duração média das faixas
}).reset_index()
```

```python
# Renomear as colunas para melhor clareza
agrupado_por_artista.columns = ['artists', 'average_popularity', 'number_of_tracks', 'average_duration']
```

```python
# Ordenar os artistas pela média de popularidade em ordem decrescente
agrupado_por_artista_ordenado = agrupado_por_artista.sort_values(by='average_popularity', ascending=False)
```

```python
# Exibir os resultados
print(agrupado_por_artista_ordenado.head(10))  # Exibir os 10 artistas mais populares, por exemplo
```

```python
agrupado_por_musica = dados_spotify.groupby('track_name')['popularity'].mean().reset_index()
```

```python
# Ordenar as musicas pela média de popularidade em ordem decrescente
agrupado_por_musica_ordenado = agrupado_por_musica.sort_values(by='popularity', ascending=False)
```

```python
# Exibir os resultados
print(agrupado_por_musica_ordenado.head(10))  # Exibir as 10 musicas mais populares, por exemplo
```
